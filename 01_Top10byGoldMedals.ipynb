```python
# 01 Top 10 Countries by Gold Medals
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


medallists = pd.read_csv('/Users/stanley/04_Developer/Python_Data_Analytics_Course-main/05_2024OlympicGames/medallists.csv')
medals_total = pd.read_csv('/Users/stanley/04_Developer/Python_Data_Analytics_Course-main/05_2024OlympicGames/medals_total.csv')
country_mapping = medallists[['country','country_code']].drop_duplicates().set_index('country_code')
top_10 = medals_total.head(10)

x = top_10['country_code'].map(country_mapping['country'])
y = top_10['Gold Medal']

colors = ['#4B0082', '#6A0DAD', '#8E24AA', '#B146C2', '#D469E8', '#E0B0FF']
cmap = mcolors.LinearSegmentedColormap.from_list('my_cmap', colors)
fig, ax = plt.subplots()

plt.bar(x,y,color=cmap(np.linspace(0, 1, len(x))), edgecolor='black', alpha=0.8)

for i, v in enumerate(y):
    plt.text(i, v + 0.3, str(v), ha='center')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.xticks(rotation = 45, ha = 'right')
plt.xlabel('Country')
plt.ylabel('Gold Medals')
plt.title('Top 10 Countries by Gold Medals')
plt.show
```
