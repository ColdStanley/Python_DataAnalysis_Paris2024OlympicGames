```python
import pandas as pd 
import matplotlib.pyplot as plt

medallists = pd.read_csv('/Users/stanley/04_Developer/Python_Data_Analytics_Course-main/05_2024OlympicGames/medallists.csv')
medals_total = pd.read_csv('/Users/stanley/04_Developer/Python_Data_Analytics_Course-main/05_2024OlympicGames/medals_total.csv')

country_mapping = medallists[['country','country_code']].drop_duplicates().set_index('country_code')

top_10_medals = medals_total.sort_values(by='Total', ascending=False).head(10)
top_10_medals_for_stacked = top_10_medals[['country_code','Gold Medal','Silver Medal','Bronze Medal']]

top_10_medals_for_stacked.plot(x='country_code', kind = 'bar', stacked=True)

plt.xticks(ticks=range(len(top_10_medals_for_stacked)), 
           labels=top_10_medals_for_stacked['country_code'].map(country_mapping['country']), 
           rotation=45, ha='right')

plt.xlabel('Country')
plt.ylabel('Medals')
plt.title('Top 10 Countries by Medals')
plt.show
```
